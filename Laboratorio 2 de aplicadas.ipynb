```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import math
```

```python
def newton(f, f_prime, x0, tol=1e-4, max_iter=100):
    for _ in range(max_iter):
        f_val = f(x0)
        df_val = f_prime(x0)
        if df_val == 0:
            return None
        x1 = x0 - f_val / df_val
        if abs(x1 - x0) < tol:
            return x1
        x0 = x1
    return None
```

$1. x^5- 6x^3+2x = 2$

```python
x = sp.symbols('x')
f_expr1 = x**5 - 6*x**3 + 2*x - 2
f1 = sp.lambdify(x, f_expr1, "numpy")
f1_prime = sp.lambdify(x, sp.diff(f_expr1, x), "numpy")

x_vals = np.linspace(-3, 3, 400)
plt.plot(x_vals, f1(x_vals))
plt.axhline(0, color='gray', linestyle='--')
plt.title("f(x) = x⁵ − 6x³ + 2x − 2")
plt.grid()
plt.show()

for x0 in [-2, -1, 0, 1, 2]:
    raiz = newton(f1, f1_prime, x0)
    if raiz:
        print(f"x0 = {x0} → raíz ≈ {raiz:.5f}, f(x) ≈ {f1(raiz):.2e}")
```

$2. senx=x^2$

```python
f_expr2 = sp.sin(x) - x**2
f2 = sp.lambdify(x, f_expr2, "numpy")
f2_prime = sp.lambdify(x, sp.diff(f_expr2, x), "numpy")

x_vals = np.linspace(-2, 2, 400)
plt.plot(x_vals, f2(x_vals))
plt.axhline(0, color='gray', linestyle='--')
plt.title("f(x) = sen(x) − x²")
plt.grid()
plt.show()

for x0 in [-1.0, 0.5, 1.0]:
    raiz = newton(f2, f2_prime, x0)
    if raiz:
        print(f"x0 = {x0} → raíz ≈ {raiz:.5f}, f(x) ≈ {f2(raiz):.2e}")
```

$3. \quad cost=t^2$

```python
t = sp.symbols('t')
f_expr3 = sp.cos(t) - t**2
f3 = sp.lambdify(t, f_expr3, "numpy")
f3_prime = sp.lambdify(t, sp.diff(f_expr3, t), "numpy")

t_vals = np.linspace(-2, 2, 400)
plt.plot(t_vals, f3(t_vals))
plt.axhline(0, color='gray', linestyle='--')
plt.title("f(t) = cos(t) − t²")
plt.grid()
plt.show()

for t0 in [0.0, 0.5, 1.0]:
    raiz = newton(f3, f3_prime, t0)
    if raiz:
        print(f"t0 = {t0} → raíz ≈ {raiz:.5f}, f(t) ≈ {f3(raiz):.2e}")
```

$4.\quad 2senx=\sqrt{x+1}$

```python
f_expr4 = 2 * sp.sin(x) - sp.sqrt(x + 1)
f4 = sp.lambdify(x, f_expr4, "numpy")
f4_prime = sp.lambdify(x, sp.diff(f_expr4, x), "numpy")

x_vals = np.linspace(-1, 2, 400)
plt.plot(x_vals, f4(x_vals))
plt.axhline(0, color='gray', linestyle='--')
plt.title("f(x) = 2sin(x) − √(x + 1)")
plt.grid()
plt.show()

for x0 in [0.0, 0.5, 1.0]:
    raiz = newton(f4, f4_prime, x0)
    if raiz:
        print(f"x0 = {x0} → raíz ≈ {raiz:.5f}, f(x) ≈ {f4(raiz):.2e}")
```

```python
def newton_sqrt3(x0, tol):
    x = x0
    while True:
        x_next = x - (x**2 - 3) / (2 * x)
        if abs(x_next - x) < tol:
            return x_next
        x = x_next

tolerance = 0.00005
initial_guess = 1.0  # No puede ser 0

# Aproximación
approx_sqrt3 = newton_sqrt3(initial_guess, tolerance)

result = 2 - approx_sqrt3

print(f"Aproximación de sqrt(3): {approx_sqrt3:.6f}")
print(f"Aproximación de 2 - sqrt(3): {result:.6f}")
```

$\frac{x}{x+1}$

```python
#1. Dominio: (−∞,−1)∪(−1,+∞) | Rango: (-∞,1)U(1,+∞)
#2. No es identificable
#3. Intersección: (0,0)
#4. Puntos criticos: No hay | creciente: siempre crece y nunca decrece
#5. No hay punto criticos
#6. No tiene concavidad
#7. Asintota vertical: x=-1 | Asintota horizontal: y=1

def f(x):
  return x/(x+1)

x1 = np.linspace(-10, -1.01, 400)
x2 = np.linspace(-0.99, 10, 400)

plt.figure(figsize=(5,4))
plt.plot(x1, f(x1), x2, f(x2), 'b', label='f(x)=x/(x+1)')
plt.scatter([0], [0], c='k', label='(0,0)')

# Asíntotas
plt.axvline(-1, color='purple', ls='--', label='x=-1')
plt.axhline(1, color='green', ls='--', label='y=1')

plt.title('Ejercicio 1')
plt.xlabel('x'); plt.ylabel('f(x)')
plt.ylim(-10,10); plt.xlim(-10,10)
plt.legend(); plt.grid(True)
plt.show()
```

$\frac{|x|+x}{2}*(3x+2)$

```python
#1. Dominio: (-∞,+∞) | Rango: (0,+∞)
#2. No es identificable
#3. Intersección: (-∞,0)
#4. Puntos criticos: x_1=-4 y x_2=0 | creciente: (0,+∞) y no decrece
#5. x_1: no es extremo | x_2: es un minimo
#6. No tiene concavidad
#7. Asintota vertical: no hay | Asintota horizontal: no hay

def g(x):
    return ((np.abs(x)+x)/2)*(3*x+2)

x = np.linspace(-5,5,600)
plt.figure(figsize=(5,4))
plt.plot(x, g(x), 'b', label='g(x)')
plt.axhline(0, color='gray', lw=0.5)

plt.scatter([0], [0], c='k', label='(0,0)')

plt.title('Ejercicio 2')
plt.xlabel('x'); plt.ylabel('g(x)')
plt.legend(); plt.grid(True)
plt.show()
```

$\frac{x^2}{x^2-1}$

```python
#1. Dominio: (-∞,+1)U(1-,1)U(1,+∞) | Rango: (-∞,+∞)
#2. No es identificable
#3. Intersección: (0,0)
#4. Puntos criticos: x=0 | creciente: no se y decrece no se
#5. x=0: es un maximo
#6. (-1,1) tiene concavidad convexa
#7. Asintota vertical: x=-1 y x=1 | Asintota horizontal: y=2

def f3(x):
  return x**2/(x**2-1)

x1 = np.linspace(-10, -1.01, 400)
x2 = np.linspace(-0.99, 0.99, 400)
x3 = np.linspace(1.01, 10, 400)

plt.figure(figsize=(5,4))
plt.plot(x1, f3(x1), 'b', x2, f3(x2), 'b', x3, f3(x3), 'b', label='f(x)')
plt.scatter([0], [0], c='k', label='(0,0)')

# Asíntotas
plt.axvline(-1, color='purple', ls='--')
plt.axvline(1, color='purple', ls='--', label='x=±1')
plt.axhline(1,  color='green', ls='--', label='y=1')

plt.title('Ejercicio 3')
plt.xlabel('x'); plt.ylabel('f(x)')
plt.ylim(-10,10); plt.xlim(-10,10)
plt.legend(); plt.grid(True)
plt.show()
```

$\frac{5.235x^3-1.245}{7.126x-3.141}$

```python
#1. Dominio: (-∞,0,44)U(0,44,+∞) | Rango: (0,+∞)
#2. No es identificable
#3. Intersección: (0,0.396)
#4. Puntos criticos: no tiene | creciente: (0.0396,+∞) y decrece (-∞,0.0396)
#5. no tiene extremos
#6. Tiene concavidad concava
#7. Asintota vertical: x=0.40 | Asintota horizontal: no tiene

def h(x):
  return (5.235*x**3 - 1.245)/(7.126*x - 3.141)

x1 = np.linspace(-10, 0.43, 400)
x2 = np.linspace(0.45, 10, 400)

plt.figure(figsize=(5,4))
plt.plot(x1, h(x1), 'b', x2, h(x2), 'b', label='h(x)')
plt.scatter([0], [h(0)], c='k', label='(0,0.396)')

x_asym = 1.245/3.141
plt.axvline(x_asym, color='purple', ls='--', label=f'x={x_asym:.2f}')

plt.title('Ejercicio 4')
plt.xlabel('x'); plt.ylabel('h(x)')
plt.legend(); plt.grid(True)
plt.show()
```

```python
x = np.linspace(-2, 2, 400)

# Función f(x) = 2x^2, que cumple:
# f es par, f(0)=0, f(1)=2
# f'(x)=4x > 0 para x>0 (crece)
# f''(x)=4 > 0 para x>0 (cóncava hacia arriba)

def f(x):
    return 2 * x**2

# Graficar
plt.plot(x, f(x), label='f(x) = 2x²')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.scatter([0, 0], [f(0), f(0)], color='red', zorder=1000)
plt.text(0, f(0)+0.2, 'Mínimo', fontsize=8)

plt.title('Bosquejo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Ejercicio III: Bosquejo de una función (grado 4) que cumpla:
# a. f es continua en todas partes
# b. f(2) = -3  y  f(6) = 1
# c. f'(2) = 0
# d. f'(6) = 3
# e. f''(6) = 0;  f''(x) > 0 para 2 < x < 6; f''(x) < 0 para x > 6

A = np.array([
    [   16,    8,   4, 2, 1],
    [ 1296,  216,  36, 6, 1],
    [   32,   12,   4, 1, 0],
    [  864,  108,  12, 1, 0],
    [  432,   36,   2, 0, 0]
], dtype=float)

B = np.array([-3, 1, 0, 3, 0], dtype=float)


a, b, c, d, e = np.linalg.solve(A, B)


def f(x):
    return a*x**4 + b*x**3 + c*x**2 + d*x + e

x = np.linspace(0, 10, 500)
plt.figure(figsize=(7,5))
plt.plot(x, f(x),   label='f(x)',   color='blue')

pts_x = [2, 6]
pts_y = [f(2), f(6)]
plt.scatter(pts_x, pts_y, color='red', zorder=5)
plt.text(2, pts_y[0]-0.5, 'f(2)=-3', fontsize=8)
plt.text(6, pts_y[1]+0.5, 'f(6)=1', fontsize=8)

plt.axvline(2, color='gray', ls='--', lw=1)
plt.axvline(6, color='gray', ls='--', lw=1)
plt.axhline(0, color='black', lw=0.5)

plt.xlim(0,10)
plt.ylim(-5,5)
plt.title('Ejercicio III')
plt.xlabel('x'); plt.ylabel('f(x)')
plt.legend(fontsize=8)
plt.grid(True)
plt.show()
```
