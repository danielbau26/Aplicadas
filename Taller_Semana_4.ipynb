```python
import numpy as np
import random
import matplotlib.pyplot as plt
import math
```

```python
#1. Serie de Fibonacci
def fibonacci(n):
  if n<2:
    return n
  return fibonacci(n-1)+fibonacci(n-2)

n=int(input("ingrese un numero: "))
x = list(range(n))
y = [fibonacci(i) for i in range(n)]

plt.plot(x,y,marker = 'o')
plt.title('n nÃºmeros de la serie de Fibonacci')
plt.xlabel('Indice')
plt.ylabel('Valor de Finonacci')
plt.grid(linestyle='--')  #cuadricula

plt.show()
```

```python
#2. AnÃ¡lisis de datos meteorolÃ³gicos

temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9,
26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

media = np.mean(temperaturas)
mediana = np.median(temperaturas)
desviacion = np.std(temperaturas)

print(f"Media: {media:.2f}Â°C")
print(f"Mediana: {mediana:.2f}Â°C")
print(f"DesviaciÃ³n estÃ¡ndar: {desviacion:.2f}Â°C")

np.random.seed(1) #si quiero que la secuencia sea la misma
temperaturas_anuales = np.random.choice(temperaturas, 365, replace=True) #genera las 364 temps en base a la lista temperaturas
temperaturas_mensuales = np.array_split(temperaturas_anuales, 12) #divide el array en los 12 meses

y=[np.median(i) for i in temperaturas_mensuales]  #la media mensual
x= ["enero","febrero","marzo","abril","mayo","junio","julio","agosto","septiembre","octubre","noviembre","diciembre"]

plt.xticks(rotation=45, fontsize=10)
plt.plot(x,y, marker = 'o')
plt.title('Temperatura media mensual')
plt.xlabel('Meses')
plt.ylabel('Temperaturas')
plt.grid(linestyle='--') #cuadricula

plt.show()
```

```python
#3. SimulaciÃ³n de un dado

np.random.seed(1)
lanzamientos = np.random.randint(1,7,1000) #numeros aleatorios del 1 al 6
valores, frecuencias = np.unique(lanzamientos, return_counts=True) #el return_counts nos dice cuantas veces se repitiÃ³ el lanzamiento

plt.bar(valores, frecuencias, color='blue')
plt.title('Frecuencia de Lanzamientos del Dado')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')

plt.show()
```

```python
#4. GrÃ¡fica de funciones trigonomÃ©tricas

x=np.linspace(0, 2*np.pi, 100)

seno = np.sin(x)
coseno = np.cos(x)
tangente=np.tan(x)

plt.plot(x, seno, label='seno')
plt.plot(x, coseno, label='coseno')
plt.plot(x, tangente, label='tangente')
plt.ylim(-15, 15) #limita Y

plt.title('Funciones TrigonomÃ©tricas')
plt.xlabel('x')
plt.ylabel('y')
plt.legend() #muestra los colores de cada funcion en la grafica
plt.grid(linestyle='--') #cuadricula

plt.show()
```

```python
#6. Pendiente de una recta

#valores cualquiera para x1,x2,y1 y y2 desde 1 al 20
x1= random.randint(1,20)
x2 = random.randint(1,20)
y1= random.randint(1,20)
y2= random.randint(1,20)

#para que en dado caso de que x1 y x2 sean iguales
while x1==x2:
  x2 = random.randint(1,20)
pendiente = (y2-y1)/(x2-x1)

def recta (x):
  rectaa = pendiente * (x-x1) +y1
  return rectaa
x=[x1,x2]
y=[y1,y2]


plt.plot(x,y)
plt.title('Grafica de pendiente de una recta')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(linestyle='--') #cuadricula

plt.show()

```

```python
#7. Movimiento parabolico

velocidadInicial = random.randint(0,20)
angulo = random.randint(1,90)

angulo_radiantes = np.radians(angulo)
total= (2*velocidadInicial*np.sin(angulo_radiantes))/9.8
tiempo = np.linspace(0, total, 100)

x = velocidadInicial * np.cos(angulo_radiantes) * tiempo
y = velocidadInicial * np.sin(angulo_radiantes) * tiempo - 0.5 * 9.8 * tiempo**2

plt.plot(x,y)
plt.title('Movimiento parabolico')
plt.xlabel('Distancia')
plt.ylabel('Altura')
plt.grid(linestyle='--') #cuadricula

plt.show()

```

```python
#9. Punto extra: fractales con python

def trianguloSierpinski(puntos, nivel):
    if nivel == 0:
        x, y = zip(*puntos)
        plt.fill(x, y, "black")
    else:
        p1, p2, p3 = puntos  # Extraer los puntos desde la lista

        medio1 = ((p1[0] + p2[0]) / 2, (p1[1] + p2[1]) / 2)
        medio2 = ((p2[0] + p3[0]) / 2, (p2[1] + p3[1]) / 2)
        medio3 = ((p1[0] + p3[0]) / 2, (p1[1] + p3[1]) / 2)

        #llamamos la funcion para que los triangulos se hagan pequeÃ±os hasta llegar a nivel == 0
        trianguloSierpinski([p1, medio1, medio3], nivel - 1)
        trianguloSierpinski([p2, medio1, medio2], nivel - 1)
        trianguloSierpinski([p3, medio2, medio3], nivel - 1)

# Puntos de un triÃ¡ngulo equilÃ¡tero
p1 = (0, 0)
p2 = (1, 0)
p3 = (0.5, np.sqrt(3)/2)

nivel = int(input("Escriba el nivel del triangulo: "))

plt.figure(figsize=(6, 6))
trianguloSierpinski([p1, p2, p3], nivel)
plt.show()
```
